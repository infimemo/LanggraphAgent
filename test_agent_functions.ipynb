```python
# USER CASE 2: USER WANT TO OPTIMIZE A PORTFOLIO - WORKING WELL
symbols = ['AAPL', 'AMZN', 'GOOG', 'TSLA', 'MSFT', 'NVDA', 'META', 'BRK-B', 'UNH', 'JPM']
def user_optimizes_portfolio(
    symbols: list,
    alpha=0.1,
    risk_aversion=1,
    mean_shrinkage=0,
    cov_shrinkage=0.1,
    min_weight=0,
    max_weight=1,
    sector_map=None,
    sector_max=None
):
    """
    Optimize portfolio weights using robust modern portfolio theory.

    Args:
        returns_df (pd.DataFrame): Asset returns with tickers as columns.
        alpha (float): Uncertainty parameter for worst-case return.
        risk_aversion (float): Risk aversion parameter in the objective.
        mean_shrinkage (float): Shrinkage intensity for mean returns.
        cov_shrinkage (float): Shrinkage intensity for covariance matrix.
        min_weight (float): Minimum weight per asset.
        max_weight (float): Maximum weight per asset.
        sector_map (dict, optional): Mapping of tickers to sectors.
        sector_max (float, optional): Maximum allocation per sector.

    Returns:
        dict: Mapping of tickers to optimal weights.
    """
    returns_df = fetch_data(symbols)
    #ortho_returns_df = orthogonalize(returns_df)
    df = optimize_portfolio(
        returns_df=returns_df,
        alpha=alpha,
        risk_aversion=risk_aversion,
        mean_shrinkage=mean_shrinkage,
        cov_shrinkage=cov_shrinkage,
        min_weight=min_weight,
        max_weight=max_weight,
        sector_map=sector_map,
        sector_max=sector_max
    )
    return df

# sample working well
optimized_port = user_optimizes_portfolio(symbols=symbols)
optimized_port
```
